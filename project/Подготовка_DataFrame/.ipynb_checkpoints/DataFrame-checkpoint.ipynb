```python
train_oil_year_accum = pd.read_csv('../../data/Преобразованные_данные/train_oil_year_accum.csv')
train_oil_year_accum['MEASURED_IN_DATE'] = pd.to_datetime(train_oil_year_accum['MEASURED_IN_DATE'])
```

```python
full_well = []
for well in train_oil_year_accum['WELL_NAME'].unique():
    
    data = train_oil_year_accum[train_oil_year_accum['WELL_NAME'] == well]
    data.set_index('MEASURED_IN_DATE', inplace=True)
    
    daily_counts = data.resample('D').count()
    if daily_counts.isnull().values.any():
        full_well.append(well)
print(f'Cкважин с не пропущенными значениями = {len(full_well)}')
```

```python
train_oil_accum_without_gaps = pd.DataFrame()
count_rows = 0


for well in train_oil_year_accum['WELL_NAME'].unique():
    
    data = train_oil_year_accum[train_oil_year_accum['WELL_NAME'] == well]
    data = data.sort_values(by='MEASURED_IN_DATE')

    df_dates = pd.DataFrame()

    df_dates['MEASURED_IN_DATE'] = data['MEASURED_IN_DATE']
    
    df_dates['NUMBER'] = list(range(count_rows, df_dates.shape[0]  + count_rows))
    count_rows += df_dates.shape[0]
    
    df_dates['DIFF_DATE'] = data['MEASURED_IN_DATE'].diff()
    df_dates = df_dates[['NUMBER', 'MEASURED_IN_DATE', 'DIFF_DATE']]

    intervals = pd.DataFrame()
    intervals['DIFF_COUNT'] = df_dates[df_dates['DIFF_DATE'] != '1 days']['NUMBER'].diff()
    intervals['INDEX_DATE'] = intervals.index
    intervals.reset_index(drop=True, inplace=True)
    intervals = intervals[['INDEX_DATE', 'DIFF_COUNT']]

    max_diff = intervals['DIFF_COUNT'].max()    
    end_max_interval = int(intervals[intervals['DIFF_COUNT'] == max_diff]['INDEX_DATE'])
    start_max_interval = int(intervals.iloc[intervals[intervals['DIFF_COUNT'] == max_diff].index - 1]['INDEX_DATE'])
    
    train_oil_accum_without_gaps = pd.concat([train_oil_accum_without_gaps, 
                                                  train_oil_year_accum.iloc[start_max_interval : end_max_interval]]) 
train_oil_accum_without_gaps.reset_index(drop=True, inplace=True)
train_oil_accum_without_gaps
```

```python
well_more_365 = train_oil_accum_without_gaps.groupby('WELL_NAME').size() > 365
list_well_more_365 = well_more_365[well_more_365].index.tolist()
train_oil_year_accum_without_gaps = train_oil_accum_without_gaps[
    train_oil_accum_without_gaps['WELL_NAME'].isin(list_well_more_365)
]

train_oil_year_accum_without_gaps.reset_index(drop=True, inplace=True)
train_oil_year_accum_without_gaps
```

```python
def make_df_with_count_days_ago(dataframe, count_days, target):
    """
    dataframe - DataFrame откуда будут браться данные.
    count_days - количество прошлых измерений (в прошлые дни), которые будут в таблице.
    target - название столбца, который будет использоваться для создания временного ряда.
    
    """
    list_df = []

    for well in dataframe['WELL_NAME'].unique():

        data = dataframe[dataframe['WELL_NAME'] == well]
        data.set_index('MEASURED_IN_DATE', inplace=True)

        target_series = data[target]

        df = pd.DataFrame()
        
        df['WELL_NAME'] = data['WELL_NAME'][count_days :].reset_index(drop=True)

        for i in range(count_days):
            df[f'{target}_{count_days - i}_days_ago'] = \
                target_series[i : target_series.count() - count_days + i].reset_index(drop=True)

        df.index = target_series.index[count_days :]
        df[f'TARGET_{target}'] = target_series[count_days :]
        list_df.append(df)
    df = pd.concat(list_df)
    return df
```
