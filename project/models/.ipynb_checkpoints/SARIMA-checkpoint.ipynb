```python
import pandas as pd
import numpy as np


from itertools import product
from tqdm import tqdm
import statsmodels.api as sm
import scipy.stats as scs

from functions.tsplot import tsplot
from functions.unboxcox import unboxcox

import seaborn as sns
import matplotlib.pyplot as plt

import warnings 
warnings.simplefilter('ignore')


%config InlineBackend.figure_formats = 'svg'
```

```python
df = pd.read_csv('../../data/preprocessed_data/train_last_50_days_ago.csv')
df['MEASURED_IN_DATE'] = pd.to_datetime(df['MEASURED_IN_DATE'])
df.set_index('MEASURED_IN_DATE', inplace=True)


test_size = 60

data_test = df[df['WELL_NAME'] == 1][['TARGET_OIL_RATE']][- test_size :]

data_train = df[df['WELL_NAME'] == 1][['TARGET_OIL_RATE']][: - test_size]
data_train['TARGET_OIL_RATE_BOXCOX'], lmbda_boxcox = scs.boxcox(data_train['TARGET_OIL_RATE']) 
data_train['TARGET_OIL_RATE_YEO_JOHNSON'], lmbda_yeo_johnson = scs.yeojohnson(data_train['TARGET_OIL_RATE'])
```

```python
plt.figure(figsize=(10, 5));
plt.grid(True)
sns.lineplot(data_train[['TARGET_OIL_RATE_BOXCOX', 'TARGET_OIL_RATE_YEO_JOHNSON']])
plt.tight_layout()
```

```python
ps = range(0, 3)
d = 1
qs = range(0, 3)
Ps = range(0, 3)
D = 1
Qs = range(0, 3)
s = 7

parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
```

```python
results = []
best_aic = np.inf

for param in tqdm(parameters_list, colour='green'):
    
    try:
        model = sm.tsa.statespace.SARIMAX(endog=data_train['TARGET_OIL_RATE'], 
                                          order=(param[0], d, param[1]), 
                                          seasonal_order=(param[2], D, param[3], s)).fit()
    except ValueError:
        print(f'wrong parameters {param}')
        continue
    aic = model.aic
    
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, aic])
    
result_table = pd.DataFrame(results, columns=['PARAMETERS', 'AIC'])
result_table.sort_values(by='AIC').head()
```

```python
best_model = sm.tsa.statespace.SARIMAX(endog=data_train['TARGET_OIL_RATE'], 
                                          order=(best_param[0], d, best_param[1]), 
                                          seasonal_order=(best_param[2], D, best_param[3], s)).fit()
print(best_model.summary())
```
